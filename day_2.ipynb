```python
import pandas as pd
from itertools import combinations
```

## Part 1

```python
data = pd.read_csv('day_2.txt', sep='\t', header=None)
```

```python
data
```

```python
data_df = pd.DataFrame({'min_value': data.min(axis=1), 'max_value': data.max(axis=1)})
data_df['diff_value'] = data_df.max_value - data_df.min_value
data_df.diff_value.sum()
```

## Part 2

```python
row_results = 0
for i in range(len(data)):
    for comb in combinations(data.iloc[i],2):
        if sorted(comb)[1] % sorted(comb)[0] == 0:
            row_results += sorted(comb)[1] / sorted(comb)[0]
            break
row_results            
```
