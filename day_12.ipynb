# Part 1

```python
with open('day_12.txt') as f:
    lines = f.readlines()
```

```python
direct_communication_dict = {}
for l in lines:
    direct_communication_dict[l.split(' ')[0]] = l.split('> ')[1][:-1].split(', ')
```

```python
zero_group = ['0']
to_collect_from = direct_communication_dict['0']

while len(to_collect_from) > 0:
    item = to_collect_from.pop(0)
    if item not in zero_group:
        zero_group.append(item)
        to_collect_from = to_collect_from + [x for x in direct_communication_dict[item] if x not in zero_group]
    print(len(to_collect_from), len(zero_group))
```

# Part 2

```python
def collect_one_group(group_member, direct_communication_dict):
    zero_group = [group_member]
    to_collect_from = direct_communication_dict[group_member]

    while len(to_collect_from) > 0:
        item = to_collect_from.pop(0)
        if item not in zero_group:
            zero_group.append(item)
            to_collect_from = to_collect_from + [x for x in direct_communication_dict[item] if x not in zero_group]
    return zero_group
```

```python
groups = {}
while len(direct_communication_dict) > 0:
    group_key = list(direct_communication_dict.keys())[0]
    group_members = collect_one_group(group_key, direct_communication_dict)
    groups[group_key] = group_members
    for member in group_members:      
        del direct_communication_dict[member]
    print(len(groups), len(direct_communication_dict))
    
```
